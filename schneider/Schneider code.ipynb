# Data Engineering & EDA

```python
def missing_data(x):
    total = x.isnull().sum().sort_values(ascending=False)
    percent = (x.isnull().sum()/x.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=["Total", "Percent"])
    return missing_data
```

```python
def correlations(x):
    numerical = x.select_dtypes(include=['int64','float64','Int64'])[:]
    plt.figure(figsize=(36,6), dpi=140)
    for j,i in enumerate(['pearson','kendall','spearman']):
        plt.subplot(1,3,j+1)
        correlation = numerical.dropna().corr(method=i)
        sns.heatmap(correlation, linewidth = 2, square=True, annot=True, cmap='BrBG')
        plt.title(i, fontsize=28)
```

```python
from sklearn.pipeline import Pipeline
```

```python
# cat_names = x.select_dtypes(exclude=["number"]).columns
# num_names = x.select_dtypes(exclude=["bool_","object_"])
cat_names = x.select_dtypes(include="object").columns
num_names = x.select_dtypes(exclude="object").columns
```

```python
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs')
ohe = OneHotEncoder()
rs = RobustScaler(with_centering=False)
```

```python
cat_trans = Pipeline(steps=[('onehot', ohe)])
```

```python
num_trans = Pipeline(steps=[('poly',PolynomialFeatures(degree = 2))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[('num', num_trans, num_names), ('cat', cat_trans, cat_names)])
```

```python
pipe = Pipeline([("preprocessing", preprocessor), ('scaler', rs), ('linear', logreg)])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 3240)
```
